#In this task, we utilized the diamonds dataset to plot the relationship between the 'carat' and 'price' attributes of diamonds with a specific clarity ('SI2') and color ('E'). We used Seaborn's scatterplot to create the chart, with the 'cut' attribute determining the colors of the points. Various modifications were made to the chart to enhance readability and visual appeal, including adjusting the figure size, font sizes, legend title, and label spacing. The resulting chart provides a clear visualization of how the carat and price of diamonds with specific characteristics correlate.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./sample_data/diamonds.csv')
```

```python
filtered_df = df[(df['clarity'] == 'SI2') & (df['color'] == 'E')]
```

```python
plt.figure(figsize=(10, 6))  # Set the figure size for better readability
sns.scatterplot(data=filtered_df, x='carat', y='price', hue='cut', palette='bright')
plt.title('Relationship Between Carat and Price for SI2 Clarity and E Color Diamonds', fontsize=14)
plt.xlabel('Carat', fontsize=12)
plt.ylabel('Price (USD)', fontsize=12)
plt.legend(title='Cut', title_fontsize='13', labelspacing=1.2)
plt.tight_layout()  # Adjust the layout to make room for the legend and axis labels
plt.show()
```
