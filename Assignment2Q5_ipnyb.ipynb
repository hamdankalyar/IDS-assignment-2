#In this task, we're visualizing heritage sites in Pakistan on a map using Folium. First, we load the data from a CSV file, specifying column names manually as the CSV doesn't have a header row. Then, we initialize a Folium map centered around the geographical center of Pakistan with an initial zoom level of 6 for a country-wide view. Next, we iterate through each row in the DataFrame representing heritage sites, adding a marker for each site to the map with its corresponding latitude and longitude coordinates. Additionally, we set up popups to display the name of each site when the marker is clicked. Finally, we display the map with all the markers representing the heritage sites in Pakistan.

```python
import folium
import pandas as pd
```

```python
# load the data
# # The CSV has no header row, so we need to specify 'header=None'
df = pd.read_csv('./sample_data/pak-heritage-sites.csv', header=None)
# provide the column names manually
df.columns = ['Latitude', 'Longitude', 'Site']
```

```python
# Initialize a folium map centered around the geographic center of Pakistan
# Latitude and longitude are set to the center of Pakistan for initial focus
# Zoom level is set to 6 which provides a good country-wide view
pakistan_map = folium.Map(location=[30.3753, 69.3451], zoom_start=6)
```

```python

# Loop through the DataFrame and add a marker for each heritage site
# 'Latitude' and 'Longitude' are used for marker positions
# 'Site' name is used for the popup that will display the site name when the marker is clicked
for index, row in df.iterrows():
    folium.Marker(
        location=[row['Latitude'], row['Longitude']],  # Position of the marker
        popup=row['Site'],  # Popup text displayed on click
        icon=folium.Icon(color='blue', icon='info-sign')  # Marker icon configuration
    ).add_to(pakistan_map)  # Add the marker to the map
```
