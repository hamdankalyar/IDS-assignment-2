#In this task, we are visualizing nuclear waste sites on a map using Folium. We first read in a dataset containing the coordinates and information about nuclear waste sites. Then, we initialize a Folium map centered on the United States. Next, we iterate over each row in the dataset, adding a marker for each nuclear waste site to the map with its corresponding latitude and longitude coordinates, along with a popup displaying additional information about the site. Finally, we display the map with all the markers representing the nuclear waste sites.

```python
!pip install folium
```

```python
import folium
import pandas as pd
```

```python
df = pd.read_csv('./sample_data/nuclear_waste_sites.csv')
```

```python
us_map = folium.Map(location=[39.8283, -98.5795], zoom_start=4, tiles='OpenStreetMap')
```

```python
for index, row in df.iterrows():
    folium.Marker([row['lat'], row['lon']], popup=row['text']).add_to(us_map)
```
